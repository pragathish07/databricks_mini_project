```python
import json

# ===== CONFIGURE YOUR NEW DATA SOURCE HERE =====
new_source = {
    "source_file": "new_data.csv",           # CSV file name in S3
    "target_table": "new_data",              # Target table name in bronze schema
    "s3_bucket": "mini-project-raw-data",    # S3 bucket name
    "s3_path": "datasets"                    # Path within bucket
}
# ===============================================

metadata_path = '/Volumes/mini_project/bronze/config_volume/ingestion_metadata.json'

with open(metadata_path, 'r') as f:
    config = json.load(f)

existing_tables = [item['target_table'] for item in config]
if new_source['target_table'] in existing_tables:
    print(f"  Table '{new_source['target_table']}' already exists in metadata!")
else:
    config.append(new_source)
    
    with open(metadata_path, 'w') as f:
        json.dump(config, f, indent=2)
    
    print(f"Added new data source: {new_source['source_file']} → {new_source['target_table']}")
    print(f"Total data sources: {len(config)}")
    print(f"\n File arrival trigger will automatically run the bronze ingestion job!")
```

```python
import json

# ===== SPECIFY TABLE TO REMOVE =====
table_to_remove = "new_data"  
# ===================================

metadata_path = '/Volumes/mini_project/bronze/config_volume/ingestion_metadata.json'

with open(metadata_path, 'r') as f:
    config = json.load(f)

original_count = len(config)

config = [item for item in config if item['target_table'] != table_to_remove]

if len(config) == original_count:
    print(f" Table '{table_to_remove}' not found in metadata")
else:
    with open(metadata_path, 'w') as f:
        json.dump(config, f, indent=2)
    
    print(f"Removed data source: {table_to_remove}")
    print(f"Remaining data sources: {len(config)}")
    print(f"\n File arrival trigger will automatically run the bronze ingestion job!")
```

```python
import json

metadata_path = '/Volumes/mini_project/bronze/config_volume/ingestion_metadata.json'


with open(metadata_path, 'r') as f:
    config = json.load(f)

print(" Validating metadata structure...\n")

errors = []
required_fields = ['source_file', 'target_table', 's3_bucket', 's3_path']

for idx, item in enumerate(config):
    
    missing_fields = [field for field in required_fields if field not in item]
    if missing_fields:
        errors.append(f"Entry {idx+1}: Missing fields {missing_fields}")
    
   
    empty_fields = [field for field in required_fields if field in item and not item[field]]
    if empty_fields:
        errors.append(f"Entry {idx+1}: Empty values in {empty_fields}")

if errors:
    print("Validation failed:\n")
    for error in errors:
        print(f"  - {error}")
else
    print(f" Validation passed!")
    print(f"All {len(config)} entries are valid")
    print(f"\n Configured tables:")
    for item in config:
        print(f"  • {item['target_table']} ← {item['source_file']}")
```
